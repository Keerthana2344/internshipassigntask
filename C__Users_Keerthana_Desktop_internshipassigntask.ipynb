```python
import requests

API_URL = "https://samples.openweathermap.org/data/2.5/forecast/hourly?q=London,us&appid=b6907d289e10d714a6e88b30761fae22"

def get_weather_data():
    response = requests.get(API_URL)
    if response.status_code == 200:
        return response.json()
    else:
        print("Error fetching weather data.")
        return None

def get_weather_by_date(data, target_date):
    for forecast in data['list']:
        if target_date in forecast['dt_txt']:
            return forecast['main']['temp']
    return None

def get_wind_speed_by_date(data, target_date):
    for forecast in data['list']:
        if target_date in forecast['dt_txt']:
            return forecast['wind']['speed']
    return None

def get_pressure_by_date(data, target_date):
    for forecast in data['list']:
        if target_date in forecast['dt_txt']:
            return forecast['main']['pressure']
    return None

def main():
    weather_data = get_weather_data()

    if weather_data:
        while True:
            print("\nMenu:")
            print("1. Get weather")
            print("2. Get Wind Speed")
            print("3. Get Pressure")
            print("0. Exit")
            choice = input("Enter your choice: ")

            if choice == '1':
                date_input = input("Enter the date (YYYY-MM-DD HH:mm:ss format): ")
                temperature = get_weather_by_date(weather_data, date_input)
                if temperature:
                    print(f"The temperature on {date_input} is {temperature} Kelvin.")
                else:
                    print("Weather data not found for the given date.")
            elif choice == '2':
                date_input = input("Enter the date (YYYY-MM-DD HH:mm:ss format): ")
                wind_speed = get_wind_speed_by_date(weather_data, date_input)
                if wind_speed:
                    print(f"The wind speed on {date_input} is {wind_speed} m/s.")
                else:
                    print("Weather data not found for the given date.")
            elif choice == '3':
                date_input = input("Enter the date (YYYY-MM-DD HH:mm:ss format): ")
                pressure = get_pressure_by_date(weather_data, date_input)
                if pressure:
                    print(f"The pressure on {date_input} is {pressure} hPa.")
                else:
                    print("Weather data not found for the given date.")
            elif choice == '0':
                print("Exiting the program.")
                break
            else:
                print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
